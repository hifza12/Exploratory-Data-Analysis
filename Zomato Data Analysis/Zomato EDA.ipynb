Here we have same column in both the data set so we will merge these two data sets on the bases of country code

```python
final_data=pd.merge(df,data ,on="Country Code", how="left")
final_data.head()
```

```python
final_data.shape
```

```python
countries_index=final_data.Country.value_counts().index
countries_index
```

```python
countries_val=final_data.Country.value_counts().values
countries_val
```

```python
plt.pie(countries_val[:3],labels=countries_index[:3],autopct='%1.2f%%')
```

```python
final_data.columns
```

```python
ratings=final_data.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:"rating count"})
ratings
```

# Observation


1. When Rating is between 4.5 to 4.9---> Excellent
2. When Rating are between 4.0 to 3.4--->very good
3. when Rating is between 3.5 to 3.9----> good
4. when Rating is between 3.0 to 3.4----> average
5. when Rating is between 2.5 to 2.9----> average
6. when Rating is between 2.0 to 2.4----> Poor

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x="Aggregate rating",y="rating count",data=ratings)
```
